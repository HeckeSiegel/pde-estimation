## Example Workflow for PDE Estimation

This provides an example workflow for a user defined equation. In a situation where we wish to analyze real data, the parameters would be different. However, here we consider an equation we analyzed in the accompanying paper.

```python
import train
```

#### Parameter and Equation Defintions

Here we write out the number of points to sample, the number of epochs, and other related parameters for the algorithm. 

We also include the equation we wish to analyze, in this case, we're looking at the Hamilton Jacobi Bellman equation. 

```python
params = {'n_epochs':30001,
          'n_points':10000,
          'n_grad_points':10000,
          'lambda_u':10,
          'lambda_f':1,
          'svd_init':500,
          'lambda_norm':1e-3,
          'width':50,
          'layers':4,
          'lr':0.002,
          'noise_levels': [1],
          'n_repetitions':1}

hjb = {'eqn_type':'hjb',
        'fcn':'x**2/(1+4*t)+1/2*log(1+4*t)',
        'domain':{'x':[-2,2],'t':[0,2]},
        'dictionary':'(u_tt,u_xx,u_t,u_x,u,u**2,u*u_x,u_x**2)',
        'err_vec':[0,-1,1,0,0,0,0,1]}
```

#### Running the algorithm

Here we simply run the train function and see how the error in the estimation of the law changes per epoch. We defined all our parameters in the above dictionaries.

```python
stats = train.train(**hjb,**params)
```

We see the final recovered vector is [-0.0089, 0.5699, -0.6084, 0.0039, 0.0000, 0.0042, 0.0000, -0.5522] with SNR of 0.5. The true vector is [0,-1,1,0,0,0,0,1] so the final error is ~3e-2.
